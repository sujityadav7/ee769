```python
import numpy as np
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.inspection import permutation_importance
```

Reading the data\
Removing columns containing large number of null values\
`dft` stores feature and `dfv` stores prediction

```python
df = pd.read_csv("https://www.ee.iitb.ac.in/~asethi/Dump/MouseTrain.csv")
rm_col = df.columns[df.isnull().sum()>50]
df.drop(rm_col, axis=1, inplace=True)
dfv = df.iloc[:,-2:]
dfv.head()
dft = df.iloc[:,:-2]
```

Converting remaining null values using iterative imputer\
Scaling the features


```python
imp = IterativeImputer(max_iter=10, random_state=0)
imp.fit(dft)
xarray = imp.transform(dft)
scaler = StandardScaler()
scaler.fit(xarray)
xarray = scaler.transform(xarray)
```

```python
dft = pd.DataFrame(xarray, columns=dft.columns)
```

HeatMap of Correlation matrix between features\
We can see some high correlation between features

```python
# plotting the heatmap
corr_matrix = dft.corr().abs()
hm = sn.heatmap(data = corr_matrix)
  
# displaying the plotted heatmap
plt.show()
```

```python
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(np.bool))
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.95)]
print(to_drop)
```

Dropping the columns

```python
dft = dft.drop(to_drop, axis=1)
```

```python
dfv.groupby(dfv.columns[0]).count()
```

Class for binary classification are unbalanced\
We will use f1 score

```python
dfv.groupby(dfv.columns[1]).count()
```

Class for multiple classification are unbalanced\
We will use F1-score as statistics

```python
dfv.iloc[:,0] = dfv[dfv.columns[0]].astype('category')
dfv.iloc[:,1] = dfv[dfv.columns[1]].astype('category')
```

```python
def drop_feature(X,Y,model,l=2):
    numfeature = dft.shape[1]
    r = permutation_importance(model, X, Y, n_repeats=10)
    impf =  r.importances_mean - l*r.importances_std > 0
    return impf
```

### Neural Network Binary

```python
mlp1 = MLPClassifier(max_iter=1000)
parameter_space = {
    'hidden_layer_sizes': [(10,),(15,),(5,)],
    'alpha': [0.05, 0.0001],
}
clf = GridSearchCV(mlp1, parameter_space, cv=5, n_jobs=-1 , scoring='f1_macro')
clf.fit(dft, dfv.iloc[:,0])
print("MLP: {}".format(clf.score(dft, dfv.iloc[:,0])))
```

```python
print('Best params:\n', clf.best_params_)
```

```python
numfeature = dft.shape[1]
r = permutation_importance(clf, dft, dfv.iloc[:,0], n_repeats=10)
plt.plot([x for x in range(numfeature)], r.importances_mean)
plt.title('Feature Importance')
plt.show()
```

```python
mlp1 = MLPClassifier(max_iter=1000)
parameter_space = {
    'hidden_layer_sizes': [(10,),(15,),(5,)],
    'alpha': [0.05, 0.0001],
}
newfn = drop_feature(dft,dfv.iloc[:,0],clf)
nclf = GridSearchCV(mlp1, parameter_space, cv=5, n_jobs=-1 , scoring='f1_macro')
nclf.fit(dft.to_numpy(copy=1)[:,newfn], dfv.iloc[:,0])
print("MLP: {}".format(nclf.score(dft.to_numpy(copy=1)[:,newfn], dfv.iloc[:,0])))
```
